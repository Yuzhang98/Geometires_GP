```python
import numpy as np
s1=np.concatenate((s1_1,s2_1,s3_1,s4_1,s5_1),axis=0)
s2=np.concatenate((s1_2,s2_2,s3_2,s4_2,s5_2),axis=0)
s10=np.concatenate((s1_10,s2_10,s3_10,s4_10,s5_10),axis=0)
```

```python
np.median(s10)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

s_10=pd.DataFrame(s10)
s_10.columns=['$a$','b']
s_10['time_Points']='10'

s_2=pd.DataFrame(s2)
s_2.columns=['$a$','b']
s_2['time_Points']='2'

s_1=pd.DataFrame(s1)
s_1.columns=['$a$','b']
s_1['time_Points']='1'
```

```python
frames = [s_1,s_2,s_10]
sample=pd.concat(frames,ignore_index=True)
Sample=sample.melt(id_vars=['time_Points'],value_vars=['$a$', 'b'],var_name='Para', value_name='value')
```

```python
from matplotlib import rcParams

fig, ax = plt.subplots(figsize=(15, 6))
sns.set_theme(style="whitegrid")

b=sns.violinplot(ax=ax,
               data=Sample,
               x='Para',
               y='value',
               hue='time_Points',
               palette='summer',
               split=False)
plt.setp(b.get_legend().get_texts(), fontsize='18') 
b.set_yticklabels(b.get_yticks(),size = 18)
plt.setp(b.get_legend().get_title(), fontsize='20') 
b.set_xticklabels(['a','b'], family='cursive',size = 18)
ax.set(xlabel='', ylabel='')
plt.ylabel('Samples-true',fontsize=18)
sns.despine()
plt.tight_layout()
plt.show()
```
